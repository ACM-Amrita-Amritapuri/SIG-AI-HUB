```python
Model1 = df[['Year', 'Present_Price','Kms_Driven', 'Owner']]
y = df["Selling_Price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(Model1,y,test_size=0.5, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
rk=LinearRegression()
rk.fit(x_train, y_train)
```

```python
predtn=rk.predict(x_test)
plt.scatter(y_test, predtn)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('LinearRegression MOdel')
a = np.polyfit(y_test, predtn, 1)
b = np.poly1d(a)
plt.plot(y_test,b(y_test), color='blue')
plt.show()
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predtn)) 
print('MSE:', metrics.mean_squared_error(y_test, predtn))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predtn))) 
```

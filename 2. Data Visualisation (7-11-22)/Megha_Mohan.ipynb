```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('/content/California_Houses.csv')
```

```python
# Print the first 5 rows of the data
df.head()
```

```python
df.info()
```

1.Plot a line graph of the total rooms and total bedrooms in the same figure and give labels as "Total Rooms" and "Total Bedrooms".

```python
sns.lineplot(data=df['Tot_Rooms'],label="Total Rooms")
sns.lineplot(data=df['Tot_Bedrooms'],label="Total Bedrooms")
plt.show()
```

2.Plot a median age Vs. median income bar graph with the values of median income lesser than 6.

```python
# Set the width and height of the figure
plt.figure(figsize=(8,6))
v=df.query('Median_Income<6').Median_Income
sns.barplot(x=df['Median_Age'],y=v)
```

3.Plot the heatmap of the correlation matrix of the dataset.

```python
plt.figure(figsize=(14,7))
sns.heatmap(data = df.corr(), annot =True)
```

4.Display a scatterplot between median income and median house value and colour encode it based on the age.

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=df['Median_Income'], y=df['Median_House_Value'], hue=df['Median_Age'])
```

5.Plot a density graph of the distance to LA and colour encode it based on the age between 30 and 40 (both included).

```python
plt.figure(figsize=(14,7))
cond_age = df.query('30 <= Median_Age <= 40').Median_Age
sns.kdeplot(data=df, x='Distance_to_LA', hue=cond_age, shade=True)
```

6.Plot any one type of graph of your choice using the plotly.express module

```python
import plotly.express as px
fig = px.scatter(df, x='Median_House_Value',y='Households', color="Population")
fig.show()
```
